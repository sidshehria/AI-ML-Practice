## 3) Perform the Below Visualizations.
### ● Univariate Analysis

```python
plt.hist(df['living area'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.title('Histogram of Living Area')
plt.show()
```

```python
sns.countplot(x='number of bedrooms', data=df, palette='viridis')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.title('Bar Chart of Number of Bedrooms')
plt.show()
```

### ● Bi - Variate Analysis

```python
plt.scatter(df['living area'], df['Price'], c='blue', alpha=0.5)
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.title('Scatter Plot of Living Area vs. Price')
plt.show()
```

```python
sns.boxplot(x='number of bedrooms', y='Price', data=df, palette='Set2')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.title('Box Plot of Number of Bedrooms vs. Price')
plt.show()
```

### ● Multivariate Analysis

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```

### 4) Perform descriptive statistics on the dataset.

```python
# Summary statistics for numeric columns
numeric_stats = df.describe()

# Count of missing values in each column
missing_values = df.isnull().sum()

# Median of 'Price' column
median_price = df['Price'].median()

# Mode of 'Number of Bedrooms' column
mode_bedrooms = df['number of bedrooms'].mode().values[0]

# Variance of 'Living Area' column
variance_living_area = df['living area'].var()

# Skewness of 'Price' column
skewness_price = df['Price'].skew()

# Kurtosis of 'Price' column
kurtosis_price = df['Price'].kurtosis()

# Print the results
print("Numeric Summary Statistics:")
print(numeric_stats)

print("\nCount of Missing Values:")
print(missing_values)

print("\nMedian Price:", median_price)
print("Mode Number of Bedrooms:", mode_bedrooms)
print("Variance of Living Area:", variance_living_area)
print("Skewness of Price:", skewness_price)
print("Kurtosis of Price:", kurtosis_price)
```

## 5) Handle the Missing values.

```python
# Replace missing values in numeric columns with the mean
numeric_cols = ['number of bedrooms', 'number of bathrooms', 'living area', 'lot area', 'number of floors', 'Built Year', 'Renovation Year', 'Lattitude', 'Longitude', 'living_area_renov', 'lot_area_renov', 'Number of schools nearby', 'Distance from the airport', 'Price']
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Replace missing values in categorical columns with the most frequent category
categorical_cols = ['waterfront present', 'condition of the house', 'Postal Code']
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

# Print the DataFrame to verify that missing values have been filled
print(df)
```
